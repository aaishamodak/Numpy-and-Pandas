```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2]}, index=['a', 'b'])
df2 = pd.DataFrame({'B': [3, 4]}, index=['b', 'c'])

s1 = pd.Series([5, 6], index=['a', 'b'], name='S1')
s2 = pd.Series([7, 8], index=['b', 'c'], name='S2')

# Concatenating Series and DataFrames vertically (row-wise)
result_vertical = pd.concat([df1, df2, s1, s2])

# Concatenating Series and DataFrames horizontally (column-wise)
result_horizontal = pd.concat([df1, df2, s1, s2], axis=1)

print("Vertical Concatenation:")
print(result_vertical)

print("\nHorizontal Concatenation:")  
print(result_horizontal)
```

```python
import pandas as pd
df1 = pd.DataFrame({'A': [1, 2]}, index=['a', 'b'])
df2 = pd.DataFrame({'B': [3, 4]}, index=['b', 'c'])

# Creating Series to concatenate
s1 = pd.Series([5, 6], index=['a', 'b'], name='S1')
s2 = pd.Series([7, 8], index=['b', 'c'], name='S2')

# Initialize an empty list to hold the concatenated rows
concat_result = []

# Iterating over rows of df1, df2, s1, and s2 using iterrows() and adding them to the list
for idx, row in df1.iterrows():
    concat_result.append(row)  # Append rows from df1

for idx, row in df2.iterrows():
    concat_result.append(row)  # Append rows from df2

for idx, row in s1.items():
    concat_result.append(pd.Series([row], index=[idx], name='S1'))  # Append rows from s1

for idx, row in s2.items():
    concat_result.append(pd.Series([row], index=[idx], name='S2'))  # Append rows from s2

# Concatenate the list of rows into a new DataFrame
final_concat = pd.concat(concat_result, axis=0)

# Displaying the result
print("Concatenated Result Using iterrows():")
print(final_concat)
```

```python
import json
files = [
    "C:\\Users\\Vaishnavi\\Downloads\\US_category_id.json",
    "C:\\Users\\Vaishnavi\\Downloads\\FR_category_id.json",
    "C:\\Users\\Vaishnavi\\Downloads\\DE_category_id.json",
    "C:\\Users\\Vaishnavi\\Downloads\\CA_category_id.json",
    "C:\\Users\\Vaishnavi\\Downloads\\GB_category_id.json"
]
# Function to load JSON data from a file
def load_json(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)
# Function to extract meaningful data (ID and Title) from category items
def extract_data(json_data):
    categories = []
    for item in json_data['items']:
        category_id = item['id']
        category_title = item['snippet']['title']
        categories.append({'id': category_id, 'title': category_title})
    return categories
# Store extracted data for each file
all_categories = {}
# Process each file
for file in files:
    country = file.split('_')[0]  # Extract country code from the filename (e.g., "US", "FR", etc.)
    json_data = load_json(file)
    categories = extract_data(json_data)
    all_categories[country] = categories
# Output the extracted data
for country, categories in all_categories.items():
    print(f"Categories for {country}:")
    for category in categories:
        print(f"ID: {category['id']}, Title: {category['title']}")
    print("\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
DF_US = pd.read_json("C:\\Users\\Vaishnavi\\Downloads\\US_category_id.json")
DF_FR = pd.read_json("C:\\Users\\Vaishnavi\\Downloads\\FR_category_id.json")
DF_DE = pd.read_json("C:\\Users\\Vaishnavi\\Downloads\\DE_category_id.json")
DF_CA = pd.read_json("C:\\Users\\Vaishnavi\\Downloads\\CA_category_id.json")
DF_GB = pd.read_json("C:\\Users\\Vaishnavi\\Downloads\\GB_category_id.json")
# Extract `id` and `title` from the JSON data for each country
def extract_categories(df):
    items = df['items']
    return [(item['id'], item['snippet']['title']) for item in items]
# Combine all categories from different countries
datasets = {
    "US": pd.DataFrame(extract_categories(DF_US), columns=["id", "title"]),
    "FR": pd.DataFrame(extract_categories(DF_FR), columns=["id", "title"]),
    "DE": pd.DataFrame(extract_categories(DF_DE), columns=["id", "title"]),
    "CA": pd.DataFrame(extract_categories(DF_CA), columns=["id", "title"]),
    "GB": pd.DataFrame(extract_categories(DF_GB), columns=["id", "title"]),
}
# Create a single DataFrame to align all categories
categories = pd.DataFrame()
for country, df in datasets.items():
    counts = df['title'].value_counts().rename(country)
    categories = pd.concat([categories, counts], axis=1)
# Replace NaN with 0 for stacking
categories = categories.fillna(0)
# Stacked Area Chart
plt.figure(figsize=(12, 8))
categories.T.plot(kind='area', stacked=True, alpha=0.8, colormap="tab20c")
plt.title('Alternative Stacked Area Chart of Video Categories', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.tight_layout()
plt.show()
```
